```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data using NumPy
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot the data
plt.plot(x, y, label='sin(x)')

# Customize the plot
plt.title('Simple Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Save the figure to a file (e.g., PNG, PDF, SVG)
plt.savefig('my_plot.png', dpi=300)  # You can specify the file format and DPI (dots per inch)

# Show the plot (optional)
plt.show()
```
