```python
#matplotlib is use to create graph between data
%matplotlib inline 
#use this line only when you are working with jupyter notebook
from matplotlib import pyplot as plt
#for to create graph we need x-axis and y-axis
x_axis = [1,2,3,5,6,7]#use as x-axis data
y_axis = [6,4,2,6,8,9]#use as y-axis data
plt.plot(x_axis,y_axis)#use to plot or to create a graph 
plt.show()#use to see the graph
```
