```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data using NumPy
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot the data
plt.plot(x, y, label='sin(x)')

# Add grid
plt.grid(True, linestyle='--', alpha=1)

# Customize the plot
plt.title('Plot with Grid')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
